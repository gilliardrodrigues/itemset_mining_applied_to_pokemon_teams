Notebook criado apenas para visualizar os sprites dos pokémons e corrigir no arquivo .csv os possíveis nomes incorretos.

```python
import pandas as pd
from IPython.display import display, HTML

pd.set_option('display.max_colwidth', None)
```

```python
def make_visible_img(url):
    return f'<img src="{url}" style="max-height:124px;"></img>'
```

```python
file_path = 'C:\\Users\\User\\Documents\\jupyter_notebooks\\ufmg\\itemset_mining_applied_to_pokemon_teams\\'
```

```python
df = pd.read_csv(file_path + 'data\\processed\\sprites_showdown.csv')
df['1º tipo'] = df['1º tipo'].apply(make_visible_img)
df['2º tipo'] = [make_visible_img(url)  if url != '-' else '' for url in df['2º tipo']]
df['Sprite'] = df['Sprite'].apply(make_visible_img)
display(HTML(df[['Sprite', 'Pokémon', '1º tipo', '2º tipo']].to_html(escape=False)))
```
